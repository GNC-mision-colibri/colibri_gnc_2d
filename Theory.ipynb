## Translation of a point mass in a 2-dimensional plane

Using the **radial and tangential components of a point mass' 2-dimensional curvilinear motion**, we know that its velocity and acceleration will be: 

- (1.1)

\begin{equation}\label{eq:v_rt}
\underrightarrow{v} = \dot{r}\hat{e}_{r} + r\dot{\theta}\hat{e}_{\theta}
\end{equation}

- (1.2)

\begin{equation}\label{eq:a_rt}
\underrightarrow{a} = (\ddot{r} - r\dot{\theta})\hat{e}_{r} + (r\ddot{\theta} + 2\dot{r}\dot{\theta})\hat{e}_{\theta}
\end{equation}

### No drag and no propulsion

#### Radial components

Considering the radial components (from Newton's second law using Earth as a fixed reference frame):

- (1.3)

\begin{align}\label{eq:F_rad}
\sum{F_r} &= ma_r \nonumber\\ 
&= m(\ddot{r} - r\dot{\theta}) \\
&= m\frac{-MG}{r^2}
\end{align}

- (1.4)

\begin{equation}\label{eq:rddot_ndp}
    \ddot{r} = r\dot{\theta}^2 - \frac{MG}{r^2}
\end{equation}

#### Tangential components
Looking at the tangential components:

- (1.5)

\begin{align}\label{eq:F_tan}
\sum{F_{\theta}} &= ma_{\theta} \nonumber\\ 
&= m(r\ddot{\theta} + 2\dot{r}\dot{\theta})\\
&= 0
\end{align}

- (1.6)

\begin{equation}\label{eq:thetaddot_ndp}
    \ddot{\theta} = -\frac{2\dot{r}\dot{\theta}}{r}
\end{equation}

### Considering drag, but no propulsion

We should first remember that we can calculate the drag force with:

- (1.7)

\begin{equation}\label{eq:Fdrag}
    \underrightarrow{F_d} = -\frac{1}{2}C_{d}A\rho\vert{v}\vert\cdot \underrightarrow{v}
\end{equation}

We can use the velocity, as stated in ***(1.1)***, to separate the drag force into radial and tangential components, knowing that the magnitude of ${v}$ is:

- (1.8)

\begin{equation}\label{eq:v_mgn}
   \vert{v}\vert = \sqrt{\dot{r}^2 + (r\dot{\theta})^2}
\end{equation}

#### Radial components:

From the radial components:

- (1.9)

\begin{align}\label{eq:F_rad_d}
\sum{F_r} &= ma_r \nonumber\\ 
&= m(\ddot{r} - r\dot{\theta}) \\
&= m\frac{-MG}{r^2} -\frac{1}{2}C_{d}A\rho\vert{v}\vert\dot{r}
\end{align}

Solving for $\ddot{r}$:

- (1.10)

\begin{align}\label{eq:rddot_dnp}
    \ddot{r} &= r\dot{\theta}^2 - \frac{MG}{r^2} - \frac{1}{2m}C_{d}A\rho\vert{v}\vert\dot{r} \nonumber \\
    &= r\dot{\theta}^2 - \frac{MG}{r^2} - \frac{1}{2m}C_{d}A\rho\sqrt{\dot{r}^2 + (r\dot{\theta})^2}\text{ } \dot{r}
\end{align}

#### Tangential components:

- (1.11)

\begin{align}\label{eq:F_tan_d}
\sum{F_{\theta}} &= ma_{\theta} \nonumber\\ 
&= m(r\ddot{\theta} + 2\dot{r}\dot{\theta})\\
&= - \frac{1}{2}C_{d}A\rho\sqrt{\dot{r}^2 + (r\dot{\theta})^2}\text{ } r\dot{\theta}
\end{align}

Solving for $\ddot{\theta}$:

- (1.12)

\begin{align}\label{eq:thetaddot_dnp}
    \ddot{\theta} &= -(2\dot{r}\dot{\theta} + \frac{1}{2m}C_{d}A\rho\sqrt{\dot{r}^2 + (r\dot{\theta})^2}\text{ } r\dot{\theta})\frac{1}{r} \nonumber\\
    &= -\dot{\theta}(\frac{2\dot{r}}{r} + \frac{1}{2m}C_{d}A\rho\sqrt{\dot{r}^2 + (r\dot{\theta})^2})
\end{align}


### With drag and propulsion

Before jumping and modifying both equations ***(1.10)*** and ***(1.12)*** we can remember first that the thrusters should only be activated once the satellite is in the right orientation, meaning that the thrust should only affect the tangential component. Therefore, we only need to modify equation ***(1.12)***

- (1.13)

\begin{align}\label{eq:F_tan_dp}
\sum{F_{\theta}} &= ma_{\theta} \nonumber\\ 
&= m(r\ddot{\theta} + 2\dot{r}\dot{\theta})\\
&= -\frac{1}{2}C_{d}A\rho\sqrt{\dot{r}^2 + (r\dot{\theta})^2}\text{ } r\dot{\theta} + F_p
\end{align}

- (1.14)

\begin{align}\label{eq:thetaddot_dp}
    \ddot{\theta} &= (-2\dot{r}\dot{\theta} - \frac{1}{2m}C_{d}A\rho\sqrt{\dot{r}^2 + (r\dot{\theta})^2}\text{ } r\dot{\theta} + \frac{F_p}{m})\frac{1}{r} \nonumber\\
    &= -\dot{\theta}(\frac{2\dot{r}}{r} + \frac{1}{2m}C_{d}A\rho\sqrt{\dot{r}^2 + (r\dot{\theta})^2}) + \frac{F_p}{mr}
\end{align}



## Single rotation of a rigid body in a 2-dimensional plane

Since the aim of this repository is to model a 3DOF rigid body orbiting the Earth (in a fixed plane), we can assume for the time being that we'll only be dealing with one rotation (around an axis normal to the fixed plane). Because of this, we have:

- (2.1)

\begin{equation}\label{eq:w_3dof}
\underrightarrow{\omega} = \omega_x\hat{i} + \omega_y\hat{j} + \omega_z\hat{k}
\end{equation}


\begin{equation*}
\omega_x = \omega_y = 0
\end{equation*}

Therefore, the sum of momentum around the rigid's body CM  ***(2.2)*** can be simplified into ***(2.3)***.

- (2.2)

\begin{equation}\label{eq:3dmoments}
\sum M_G=(\dot{H}_G)_{G_{xyz}} + \Omega \times H_G 
\end{equation}



- (2.3)

\begin{align}\label{eq:sumMz}
\sum\underrightarrow{M} &= I_z \dot{\omega_z} \nonumber\\
&= I_z \alpha_z
\end{align}
